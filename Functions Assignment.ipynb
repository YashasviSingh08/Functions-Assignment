```python
# Q 6. Explain the concept of generators in Python and how they are defined?
# Ans 6.A generator is a function that gives values one by one using `yield`.  
# It does not store all values in memory.  

# Example:
def my_generator():
    for i in range(1,3):
        yield i
gen = my_generator()
print(next(gen)) 
print(next(gen)) 
```

```python
# Q 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
# Ans 10. map(function, iterable): Applies a function to each item.  
# filter(function, iterable): Returns elements for which the function returns True.  
# reduce(function, iterable): Reduces the iterable to a single value by applying the function cumulatively.  

# Example:
from functools import reduce

nums = [1, 2, 3, 4, 5]

squared = list(map(lambda x: x**2, nums))
print(squared)
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
product = reduce(lambda x, y: x * y, nums) 
print(product)
```

```python
# Q 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
# Ans 4.

def is_prime(n):
    if n < 2:
        return False  
    for i in range(2, n):  
        if n % i == 0:
            return False 
    return True  

prime = [num for num in range(1, 201) if is_prime(num)]
print(prime)
    
```

```python
# Q 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
#      terms.

# Ans 5.
class Fibonacci:
    def __init__(self, n):
        self.n = n 

    def __iter__(self):
        a, b = 0, 1
        for i in range(self.n):
            yield a
            a, b = b, a + b

n_terms = int(input("Enter the number of Fibonacci terms: "))
fib_iterator = Fibonacci(n_terms)

print("Fibonacci Sequence:")
for num in fib_iterator:
    print(num, end=" ")
```

```python
# Q 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
# Ans 6.
def power_two(max):
    for i in range(max + 1):
        yield 2 ** i

max_exp = int(input("Enter the maximum exponent: "))

print(f"Powers of 2 up to {max_exp}:")
for i in power_two(max_exp):
    print(i, end=" ")
```

```python
# Q 7. Implement a generator function that reads a file line by line and yields each line as a string.
# Ans 7. 
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            line = file.readline()
            while line:
                yield line.strip()
                line = file.readline()
    except FileNotFoundError:
        print("File not found. Please check the file path.")

file_path = input("Enter the file path: ")
line_generator = read_file_line_by_line(file_path)

print("\nFile contents:")
for line in line_generator:
    print(line)
```

```python
# Q 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
# Ans 9. 

def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

celsius_temps = [0, 20, 30, 37, 100]
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print("Temperatures in Fahrenheit:", fahrenheit_temps)
```

```python
# Q 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
# Ans 10. 

def remove_vowels(string):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda char: char not in vowels, string))

input_string = input("Enter a string: ")
result = remove_vowels(input_string)

print("String without vowels:", result)
```

```python
def remove_vowels(string):
    vowels = "aeiouAEIOU"
    for i in string:
        if i in vowels:
            string = string.replace(i,"")
        
    return string

input_string = input("Enter a string: ")
result = remove_vowels(input_string)

print("String without vowels:", result)
```
